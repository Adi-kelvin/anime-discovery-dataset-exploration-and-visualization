```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib as plt
import kaggle
import zipfile
from kaggle.api.kaggle_api_extended import KaggleApi
```

```python
#authenticating access to kaggle 
kaggle_api = KaggleApi()
kaggle_api.authenticate()
```

```python
# downloading dataset file form kaggle
kaggle_api.dataset_download_file('dsfelix/animes-dataset-2023','anime-exploratory-dataset-2023.csv')
```

```python
#extracting zip file
with zipfile.ZipFile('anime-exploratory-dataset-2023.csv.zip','r') as zif:
    zif.extractall()
```

```python
#loading and viewing data using pandas
data_frame=pd.read_csv(r'anime-exploratory-dataset-2023.csv')
data_frame
```

```python
#dropping duplicates(if there are any)
data_frame=data_frame.drop_duplicates()
data_frame
```

```python
#dropping un-need columns
#dropping all columns with the prefix genre_
data_frame=data_frame.drop(columns=[i for i in data_frame if 'genre_' in i])
#dropping un-needed columns and ambiguous columns
data_frame = data_frame.drop(columns=['synopsis','status','scored_by','members','image_url','favorites'])
data_frame
```

```python
#cleaning 'duration' column
data_frame['duration']=data_frame['duration'].str.rstrip('per ep')
data_frame['duration']=data_frame['duration'].str.rstrip('min')
data_frame['duration']= data_frame['duration'].str.replace(' ','').astype(str)
data_frame['duration'] = data_frame['duration'].str.replace('hr','').astype(str)
data_frame['duration'] = data_frame['duration'].str.replace('h','').astype(str)
data_frame['duration'] = data_frame['duration'].str.replace('Unknown','NaN').astype(str)
data_frame['duration'] = data_frame['duration'].str.replace('Unknow','NaN').astype(str)
data_frame['duration'] = data_frame['duration'].str.replace('nan','NaN')
#standardizing duration to minutes
def duration_conversion(x): #calculating minute from hh mm ss format
    if (len(x)>2 and x.isdigit()):# digits with a length greater than 2
        h=x[:(len(x)-2)]#the hour string
        m=x[(len(x)-2):]#the minute string
        #casting as int
        h =int(h)
        m=int(m)
        duration = (h*60) + m #calculate minute equivalent
        return duration#return minute
    elif('sec' in x): #string with 'sec' postfix
        x=x.rstrip('sec') #remove 'sec' postfix
        x=float(x)#cast as float
        sec=x/60.0#calculate minute equivalent
        return sec#return minute
    else:#else return string
        return x
    
data_frame['duration'] = data_frame['duration'].apply(lambda x: duration_conversion(x))#iterate through column 'duration' and apply conversion
# Change column type to float for column: 'duration'
data_frame = data_frame.astype({'duration': 'float'})
data_frame#print
```

```python
# Sort by column: 'popularity' (ascending)
data_frame = data_frame.sort_values(['popularity'])
#creating a data frame for rating and popularity
grouped_rating_data_frame = pd.DataFrame({'Ratings':data_frame['rating'],'Popularity':data_frame['popularity']})
#grouping by Rating
grouped_rating_data_frame=grouped_rating_data_frame.groupby(['Ratings']).sum()#group data by 'Rating' by summing corresponding rows
grouped_rating_data_frame.plot(kind='bar',subplots=True,figsize=(19,19))#plot a bar chart using grouped_rating_data_frame
```

```python
#creating sub data frame for type and popularity
grouped_type_data_frame = pd.DataFrame({'Types':data_frame['type'],'Popularity':data_frame['popularity']})
#grouping by Rating
grouped_type_data_frame=grouped_type_data_frame.groupby(['Types']).sum()#group data by 'Types' by summing corresponding rows
grouped_type_data_frame.plot(kind='bar',subplots=True,figsize=(20,19),)#plot a bar chart using grouped_rating_data_frame
```

```python
#save data frame into a new file
data_frame.to_csv('custom-anime-exploratory-dataset-2023.csv',index=False)
```
